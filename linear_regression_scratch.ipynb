```python
# imports
import matplotlib.pyplot as plt
import random
import numpy as np

# generate data -- 100 coordinate pairs of the form (x, y)
x = np.array([random.uniform(0, 100) for _ in range(100)])
y = np.array([element * random.uniform(1, 4) for element in x])

# gradient descent function for optimization
def gradient_descent(slope_curr, intercept_curr, x, y, learning_rate):
    N = len(x)
    slope_grad = 2/N * sum(-x * (y - (slope_curr * x + intercept_curr)))
    intercept_grad = 2/N * sum(-1 * (y - (slope_curr * x + intercept_curr)))

    slope = slope_curr - slope_grad * learning_rate
    intercept = intercept_curr - intercept_grad * learning_rate

    return slope, intercept

# retrieve best fit + plot
slope = 0
intercept = 0
epochs = 300
learning_rate = 0.0001

for epoch in range(epochs):
    slope, intercept = gradient_descent(slope, intercept, x, y, learning_rate)

print("Slope:", slope)
print("Intercept:", intercept)

plt.scatter(x, y, color='Black')
plt.plot(x, slope*x + intercept, color='Blue')
plt.show()
```
