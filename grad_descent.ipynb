```python
## function y = x^2 ; y' = 2x
## gradient descent with momentum
import matplotlib.pyplot as plt
import numpy as np

epochs = 100
lr = 0.001
curr_c = (100, 100**2)
x_plot = np.arange(-100, 100)
v = 0

for _ in range(epochs):
    v = v*0.9 + lr * 2 * curr_c[0]
    x = curr_c[0] - v
    y = x**2
    curr_c = (x, y)

    plt.plot(x_plot, x_plot**2)
    plt.scatter(curr_c[0], curr_c[1], marker='*', s=300, color='Purple')
    plt.pause(0.0001)
    plt.clf()
```

```python
## function y = x^3 ; y' = 3x^2
import matplotlib.pyplot as plt
import numpy as np

epochs = 200
lr = 0.0001
curr_c = (100, 100**3)
x_plot = np.arange(-100, 100)

for _ in range(epochs):
    x = curr_c[0] - lr * 3 * curr_c[0]**2
    y = x**3
    curr_c = (x, y)

    plt.plot(x_plot, x_plot**3)
    plt.scatter(curr_c[0], curr_c[1], marker='*', s=300, color='Purple')
    plt.pause(0.0001)
    plt.clf()
```

```python
## z = x^2 + y^2
import numpy as np
import matplotlib.pyplot as plt

def find_Z(x, y):
    return x**2 + y**2

def gradient(x, y):
    return 2*x, 2*y

x = np.arange(-5, 5, 0.05)
y = np.arange(-5, 5, 0.05)

X, Y = np.meshgrid(x, y)
Z = find_Z(X, Y)

lr = 0.1
epochs = 50
curr_c = (-5, 5, find_Z(-5, 5))

for _ in range(epochs):
    x_grad, y_grad = gradient(curr_c[0], curr_c[1])

    x = curr_c[0] - lr * x_grad
    y = curr_c[1] - lr * y_grad
    z = find_Z(x, y)

    curr_c = (x, y, z)

    ax = plt.subplot(projection="3d", computed_zorder = False)
    ax.plot_surface(X, Y, Z, cmap="Blues", zorder=0)
    ax.scatter(curr_c[0], curr_c[1], curr_c[2], color='red', marker='*', s=450, zorder=1)
    plt.pause(0.0001)
    ax.clear()
```
